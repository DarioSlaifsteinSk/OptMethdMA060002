### 5.
Given that $A\succ 0 \in \mathbb{R}^{n \times n}$ and $x,\ b,\ d \in \mathbb{R}^n$ we need to find the optimal $\gamma^*$ so that:
\begin{equation}
f(\gamma)=(A(x+\gamma d),x+\gamma d)+(b, x+\gamma d) \rightarrow \min_{f \in \mathbb{R}}
\end{equation}
The optimal-step size method can be summed up as:
\begin{equation}
x^{k+1}=x^k - \gamma.\nabla f^k
\end{equation}
from our initial $f(\gamma)$ we can assume a generic quadratic $f(x)$ as:
\begin{equation}
f(x)=x^T A x+b^T x
\end{equation}
the gradient of this function is:
\begin{equation}
\nabla f(x)= g = 2A x+ b
\end{equation}
From now on, just for the sake of clarity, we will explicitly write the superscripts $k$ indicating the number of the iteration.
Now, to find $\gamma^*$ we need to find the optimal $f(x^{k+1})$ with respect to our step-size. Hence:
\begin{equation}
0 = \frac{\partial f^{k+1}}{\partial \gamma}=\frac{\partial f(x^{k+1})}{\partial \gamma}
\end{equation}
\begin{equation}
0 = \frac{\partial f(x^k - \gamma^* g^k)}{\partial \gamma}
\end{equation}
\begin{equation}
0 = \frac{\partial f(x^k - \gamma^* (2A x^k+ b))}{\partial \gamma}
\end{equation}
\begin{equation}
\frac{\partial}{\partial \gamma}(f(x^k - \gamma^* g^k)) = \frac{\partial}{\partial \gamma}((x^k - \gamma^* g^k)^T.A.(x^k - \gamma^* g^k)+b^T.(x^k - \gamma^* g^k))
\end{equation}

Which, if we looks closely its actually an inner product with gradient in this iteration $g^k$
\begin{equation}
0=(2A(x^k-\gamma.g^k)+b,-g^k)
\end{equation}
\begin{equation}
0=-(g^k,g^k)+\gamma(2A.g^k,g^k)
\end{equation}
\begin{equation}
(g^k,g^k)=\gamma(2A.g^k,g^k)
\end{equation}

\begin{equation}
\gamma=\frac{(g^k,g^k)}{(2A.g^k,g^k)}
\end{equation}

We need to show that if for the first step we reach the minimum, so $\nabla f(x^1)=0$:
\begin{equation}
\nabla f(x^1)=0=Q x^1 - b = Q.(x^0 - \gamma \nabla f^0) - b

\begin{equation}
\nabla f(x^1)=0=Q x^1 - b = Q.(x^0 - \gamma (Q.x^0-b)) - b

So, we take the left side and solve it in terms of x^0:
\begin{equation}
0 = Q.(x^0 - \gamma (Q.x^0-b)) - b

0 = (Q.x^0 - \gamma Q.(Q.x^0-b)) - b = Q.x^0 - \gamma Q.x^0 - Q \gamma.b - b
\end{equation}
\begin{equation}
0=(I-\gamma.Q)(Q.x^0-b)
\end{equation}

Hence, $g^0=Q.x^0-b$ is an eigenvector of $Q$.

The gradient descent method has the following function:

```python
def bisection(f,x,grad,a,b,N):
    # initial values
    an = a; bn = b;
    cn = (an + bn)/2
    # evaluate first set
    fa=f(x[0]-an*grad[0],x[1]-an*grad[1]);
    fb=f(x[0]-bn*grad[0],x[1]-bn*grad[1]);
    fc=f(x[0]-cn*grad[0],x[1]-cn*grad[1]);
    for n in range(1,N+1): # loop through maxIter
        # mid points
        dn=(an+cn)/2; en=(bn+cn)/2;
        fd=f(x[0]-dn*grad[0],x[1]-dn*grad[1]);
        fe=f(x[0]-en*grad[0],x[1]-en*grad[1]);
        
        if fd < fc: # move to the right
            an = an; bn = cn; cn = dn;
        elif fe > fc: # move to the left
            an = cn; cn = en; bn = bn;
        elif fc < np.minimum(fd, fe): # shrink interval
            an = dn; cn = cn; bn = en;
    return (an + bn)/2;

def gradDescent(x0,f,gradF,alpha0,steep, maxIter=50): # main function
    ## Initial values
    x=x0 #seed
    grad=gradF(x[0][0],x[0][1])
    alpha=alpha0
    
    # Parameters
    err=1
    tol=1e-5
    k=0

    while err > tol and k < maxIter: # iterations
        k=k+1 

        # Step 1: pick alpha
        if steep == 1: # check for scalar step-size
            # Steepest descent
            a = 0.005 #lower bound for alpha
            b = 5 #upper bound 
            N = 100 # number of iterations
            alphak = bisection(f,x[k-1],grad[k-1],a,b,N) # doesn't work
            alpha = np.append(alpha, [alphak]) # find the optimum alpha
        else:
            alpha = np.append(alpha, alpha0)
        
        # Step 3: calculate x for k+1
        x = np.append(x, [x[k-1] - alpha[k-1]*grad[k-1]],axis=0) # x^k+1
        # Step 1: Calculate gradient
        grad = np.append(grad,gradF(x[k][0],x[k][1]),axis=0)
        # Step 4: convergence
        err=np.abs(f(x[k][0],x[k][1])-f(x[k-1][0],x[k-1][1])) # f[k]-f[k-1]
    return x,k;
```

Just as a proof of concept we try to minimize the function for a given seed $x_0=(1,10)$

```python
def f(x):
    n=np.shape(x)
    secondTerm=0
    for i in range(n,2):
        sumTerm = 2*x[i-1]**2-x[i]-1
        secondTerm = secondTerm+(sumTerm)**2
    return ((x[0]-1)**2)/4+secondTerm;
```

Its gradient is:
\begin{equation}
\frac{\partial f}{\partial x_n}=-2.(2x_{n-1}^2-x_n-1)+8.x_n.(2x_n^2-x_{n+1}-1)
\end{equation}

```python
def gradF(x):
    n=np.shape(x);
    parDevF=np.zeros(n)
    for i in range(n[0]):
        if i == (n[0]-1):
            parDevF[i]=np.array(-2*(2*x[i-1]**2-x[i]-1)+8*x[i]*(2*x[i]**2-1))
        else:
            parDevF[i]=np.array(-2*(2*x[i-1]**2-x[i]-1)+8*x[i]*(2*x[i]**2-x[i+1]-1))
    return parDevF;
```
